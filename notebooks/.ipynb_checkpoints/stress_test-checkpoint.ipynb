```python
from mytorch import FloatTensor as FT, CPU, GPU
```

# Some simple stats experiments

Find the max of 1000 samples from the normal distribution... and repeat the experiment 10 times

```python
x = FT.randn((10, 1000), CPU)
y = x.max(1)
y
```

Do the experiment again, on GPU

```python
x = FT.randn((10, 1000), GPU)
y = x.max(1)
y.move_to_device(CPU)
y
```

Sample the normal distribution 1000 times, and take sample mean and variance.  Repeat the experiment 10x.

```python
x = FT.randn((10, 1000), CPU)
means = x.mean(1)
vs = x.var(1)      # avoid the keyword 'vars'
print("Means:")
print(means)
print("Variances:")
print(vs)
```
